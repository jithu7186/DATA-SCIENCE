```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2=pd.DataFrame({'rkey':['kee','bar','baz','fij'],
                  'values':[5,6,7,8]})
```
