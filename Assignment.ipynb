## 1. What is the distribution of users across different course categories?


```python
# Set plot style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.countplot(data=online, x='CourseCategory', palette='viridis',hue='CourseCategory')
plt.title('Distribution of Users Across Different Course Categories')
plt.xlabel('Course Category')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()


```

## 2. What is the average time spent on courses for each category?


```python

plt.figure(figsize=(12, 6))
avg_time_spent = online.groupby('CourseCategory')['TimeSpentOnCourse'].mean().sort_values()
sns.barplot(x=avg_time_spent.index, y=avg_time_spent.values, palette='viridis')
plt.title('Average Time Spent on Courses for Each Category')
plt.xlabel('Course Category')
plt.ylabel('Average Time Spent (hours)')
plt.xticks(rotation=45)
plt.show()
```

## 3.Correlation between the number of videos watched and course completion

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=online, x='CourseCompletion', y='NumberOfVideosWatched', palette='viridis')
plt.title('Correlation Between Number of Videos Watched and Course Completion')
plt.xlabel('Course Completion (0: No, 1: Yes)')
plt.ylabel('Number of Videos Watched')
plt.xticks([0, 1], ['Not Completed', 'Completed'])
plt.show()
```

## 4.Average quiz score for completed vs non-completed courses

```python
# Question 
plt.figure(figsize=(12, 6))
sns.boxplot(data=online, x='CourseCompletion', y='QuizScores', palette='viridis')
plt.title('Average Quiz Score for Completed vs Non-Completed Courses')
plt.xlabel('Course Completion (0: No, 1: Yes)')
plt.ylabel('Average Quiz Score')
plt.xticks([0, 1], ['Not Completed', 'Completed'])
plt.show()
```

## 5. Device type effect on course completion rates

```python

plt.figure(figsize=(12, 6))
completion_by_device = online.groupby('DeviceType')['CourseCompletion'].mean()
sns.barplot(x=completion_by_device.index, y=completion_by_device.values, palette='viridis')
plt.title('Course Completion Rates by Device Type')
plt.xlabel('Device Type (0: Desktop, 1: Mobile)')
plt.ylabel('Course Completion Rate')
plt.xticks([0, 1], ['Desktop', 'Mobile'])
plt.show()
```

## 6. Relationship between the number of quizzes taken and the course completion rate

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=online, x='CourseCompletion', y='NumberOfQuizzesTaken', palette='viridis')
plt.title('Number of Quizzes Taken vs. Course Completion')
plt.xlabel('Course Completion (0: No, 1: Yes)')
plt.ylabel('Number of Quizzes Taken')
plt.xticks([0, 1], ['Not Completed', 'Completed'])
plt.show()


```

## 7. Completion rate variation across different course categories

```python

plt.figure(figsize=(12, 6))
completion_rate_by_category = online.groupby('CourseCategory')['CompletionRate'].mean().sort_values()
sns.barplot(x=completion_rate_by_category.index, y=completion_rate_by_category.values, palette='viridis')
plt.title('Completion Rate Across Different Course Categories')
plt.xlabel('Course Category')
plt.ylabel('Average Completion Rate')
plt.xticks(rotation=45)
plt.show()
```

## 8. Average number of quizzes taken across different course categories

```python

plt.figure(figsize=(12, 6))
avg_quizzes_taken = online.groupby('CourseCategory')['NumberOfQuizzesTaken'].mean().sort_values()
sns.barplot(x=avg_quizzes_taken.index, y=avg_quizzes_taken.values, palette='viridis')
plt.title('Average Number of Quizzes Taken Across Different Course Categories')
plt.xlabel('Course Category')
plt.ylabel('Average Number of Quizzes Taken')
plt.xticks(rotation=45)
plt.show()
```

## 9. How does the device type affect the number of quizzes taken?

```python
plt.figure(figsize=(12, 6))
quizzes_by_device = online.groupby('DeviceType')['NumberOfQuizzesTaken'].mean()
sns.barplot(x=quizzes_by_device.index, y=quizzes_by_device.values, palette='viridis')
plt.title('Number of Quizzes Taken Across Different Device Types')
plt.xlabel('Device Type (0: Desktop, 1: Mobile)')
plt.ylabel('Average Number of Quizzes Taken')
plt.xticks([0, 1], ['Desktop', 'Mobile'])
plt.show()
```

## 10. How does the completion rate vary with the time spent on the course?

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=online, x='TimeSpentOnCourse', y='CompletionRate', palette='viridis')
plt.title('Completion Rate vs. Time Spent on Course')
plt.xlabel('Time Spent on Course (hours)')
plt.ylabel('Completion Rate')
plt.show()
```
