#### ONE SAMPLE T TEST

```python
from scipy.stats import ttest_1samp


sample_data=[65,66,62,68,61,67]

# Hypothesis Testing

t_stat,p_value=ttest_1samp(sample_data,65)
alpha=0.05
if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail To Reject Null Hypothesis")

print(t_stat)
print(p_value)
```

```python
from statsmodels.stats.proportion import proportions_ztest 


success=120
total_trials=200

#Hypothesis Testing

z_stat,p_value=proportions_ztest(success,total_trials,value=0.75)

alpha=0.05

if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail To Reject Null Hypothesis")
```

#### TWO SAMPLE T TEST

```python
from scipy.stats import ttest_ind

group_a=[10,20,55,60,75]
group_b=[30,15,48,52,66]

t_stat,p_value=ttest_ind(group_a,group_b)

alpha=0.05

if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail To Reject Null Hypothesis")
```

#### TWO SAMPLE PROPORTION TEST

```python
from statsmodels.stats.proportion import proportions_ztest

success_a,trial_a=50,100
success_b,trial_b=10,100

#Hypothesis Testing

z_stat,p_value=proportions_ztest([success_a,success_b],[trial_a,trial_b],value=0.75)

alpha=0.05

if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail To Reject Null Hypothesis")
```

#### PAIRED T TEST

```python
from scipy.stats import ttest_rel

bp_before=[120,110,130,160,125]
bp_after=[130,150,148,152,126]

t_stat,p_value=ttest_rel(bp_before,bp_after)

alpha=0.05

if p_value<alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail To Reject Null Hypothesis")
```
